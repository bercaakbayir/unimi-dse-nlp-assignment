```python
import pandas as pd
import json
import os
from IPython.display import display
```

```python
pd.set_option('display.max_colwidth', None)
```

```python


def load_data_to_df(file_path, k):
    if not os.path.exists(file_path):
        print(f"File not found at: {os.path.abspath(file_path)}")
        return None

    # Check if it's a JSONL file
    if file_path.endswith('.jsonl'):
        df = pd.read_json(file_path, lines=True)
    else:
        df = pd.read_json(file_path)

    print("Columns in Dataset:", df.columns.tolist())
    print("\nFirst 3 rows:")
    # Using 'head' safely without assuming columns exist
    display(df.head(k))

    return df
```

```python
hotpotqa_path = './data/hotpotqa/hotpot_dev_distractor_v1.json'
fever_path = './data/fever/train.jsonl'
```

```python
df_hotpotqa = load_data_to_df(hotpotqa_path,k=3)
```
